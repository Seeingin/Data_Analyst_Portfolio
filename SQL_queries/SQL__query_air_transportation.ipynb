## Air Transportation
Description of the demonstration database and connection to the database

image.png

1. Table bookings.aircrafts Each aircraft model is identified by its three-digit code (aircraft_code). The model name (model) and the maximum flight range in kilometers (range) are also indicated.

2. Table bookings.airports Airport is identified by a three-letter code (airport_code) and has its own name (airport_name). The city is not a separate entity, but the city is given and may serve to define airports of the same city. Longitude, latitude, and timezone are also specified.

3. Table bookings.boarding_passes When checking in for a flight, which is possible one day before the scheduled departure date, the passenger is issued a boarding pass. It is identified in the same way as the flight - with the ticket number and flight number. Boarding passes are assigned consecutive numbers (boarding_no) in the order of check-in for the flight (this number will be unique only within the flight). The boarding pass contains the seat number (seat_no).

4. Table bookings.bookings The passenger makes a reservation in advance (book_date, maximum one month before the flight) for himself and possibly several other passengers. The reservation is identified by a number (book_ref, a six digit combination of letters and numbers). The total_amount field stores the total cost of all passengers' flights included in the reservation.

5. Table bookings.flights A flight always connects two points - departure_airport and arrival_airport. There is no such thing as a "connecting flight". Each flight has a scheduled date and time of departure (scheduled_departure) and arrival (scheduled_arrival). Actual_departure and actual_arrival times may differ: usually not much, but sometimes by several hours if the flight is delayed.

6. Table bookings.seats The seats define the cabin layout of each model. Each seat is identified by its number (seat_no) and has a assigned class of service (fare_conditions) - Economy, Comfort or Business.

7. Table bookings.tickets The ticket has a unique number (ticket_no) consisting of 13 digits. The ticket contains passenger's identifier (passenger_id) - number of the identity document, his name and surname (passenger_name) and contact information (contact_date). Neither passenger_id nor name are permanent (it is possible to change passport, it is possible to change name), therefore it is impossible to find all tickets of one and the same passenger unambiguously.

8. Materialized view bookings.routes Table of flights contains redundancy: you could extract information about the route (flight number, departure and destination airports) from it, which does not depend on specific flight dates.

## Let's answer the following questions:

1. Which cities have more than one airport?

```python
%%sql

select a.airport_code as code, a.airport_name, a.city
from airports a
where a.city in (
 	select aa.city
	from airports aa
 	group by aa.city
 	having count(*) > 1
 )
order by a.city, a.airport_code;
```

2. Which airports have flights operated by an airplane with a maximum range? Using a subquery

```python
%%sql

select distinct a2.airport_name
from airports a2
join flights f  on  a2.airport_code = f.departure_airport
where f.aircraft_code in (
	select aircraft_code
	from aircrafts a 
	group by aircraft_code, a.range 
	order by "range" desc 
	limit 1)
group by a2.airport_name 
order by a2.airport_name;
```

3. Output the 10 flights with the maximum departure delay using the LIMIT operator

```python
%%sql

select flight_no, flight_id, actual_departure - scheduled_departure as "delay time"
from flights f 
where actual_departure - scheduled_departure is not null
order by "delay time" desc
limit 10;
```

4. Were there any reservations for which boarding passes were not received?

```python
%%sql

select count(bookings.book_ref)
from bookings
full outer join tickets on bookings.book_ref = tickets.book_ref
full outer join boarding_passes on boarding_passes.ticket_no = tickets.ticket_no
where boarding_passes.boarding_no is null;
```

5. Let's find the number of empty seats for each flight, their % ratio to the total number of seats in the plane.
Let's add a column with cumulative total - total accumulation of the number of departed passengers from each airport for each day, i.e. this column should reflect cumulative sum - how many people have already departed from this airport on this or earlier flights during the day.

```python
%%sql

select f.flight_id as "id ", 
	f.aircraft_code as " ", 
	f.departure_airport as " ", 
	date(f.actual_departure) as " ",
	(s.count_seats - bp.count_bp) as " ",
	round(((s.count_seats - bp.count_bp) * 100. / s.count_seats), 2) as "%   - ",
	sum(bp.count_bp) over (partition by date(f.actual_departure), f.departure_airport order by f.actual_departure) as "",
	bp.count_bp as "-  "
from flights f
join (
	select bp.flight_id, count(bp.seat_no) as count_bp
	from boarding_passes bp
	group by bp.flight_id) as bp on bp.flight_id = f.flight_id 
join (
	select s.aircraft_code, count(s.seat_no) as count_seats
	from seats s 
	group by s.aircraft_code) as s on f.aircraft_code = s.aircraft_code
where f.actual_departure is not null
order by date(f.actual_departure)
limit 10;
```

6. Let's find the percentage of flights by type of aircraft from the total number.

```python
%%sql

select t.model, t."total flights", t."count by model", t."% of flights"
from(
	select a.model,  
	count(f.flight_id) over () as "total flights", 
	count(f.flight_id) over (partition by f.aircraft_code) as "count by model",
	(count(f.flight_id) over (partition by f.aircraft_code) * 100) / count(f.flight_id) over () as "% of flights"
	from flights f 
	left join aircrafts a on f.aircraft_code = a.aircraft_code
	group by a.model, f.aircraft_code, f.flight_id
) t
group by t.model, t."total flights", t."count by model", t."% of flights";
```

7. Were there cities that you could get to in business class cheaper than economy class as part of the flight?

```python
%%sql

with econom as
	(select flight_id, max(amount)
	from ticket_flights
	where fare_conditions = 'Economy'
	group by flight_id),
business as
	(select flight_id, min(amount) as min
	from ticket_flights
	where fare_conditions = 'Business' 
	group by flight_id)
select e.flight_id, min, max, a1.city, a2.city
from econom e
join business b on e.flight_id = b.flight_id
left join flights f on e.flight_id = f.flight_id and b.flight_id = f.flight_id
left join airports a1 on a1.airport_code = f.arrival_airport
left join airports a2 on a2.airport_code = f.departure_airport
where max > min;


select fv.departure_city, fv.arrival_city
from (
	select flight_id
	from ticket_flights
	group by flight_id
	having max(amount) filter (where fare_conditions = 'Economy') > min(amount) filter (where fare_conditions = 'Business')) t 
join flights_v fv on fv.flight_id = t.flight_id;
```

8. What cities have no direct flights between them? 

```python
%%sql

with cte as(
	select distinct a.city as departure_city , b.city as arrival_city, a.city||'-'||b.city as route 
	from airports as a, (select city from airports) as b
	where a.city != b.city
	--where a.city > b.city     
	order by route),
cte2 as( 
	select distinct a.city as departure_city, aa.city as arrival_city, a.city||'-'|| aa.city as route  
	from flights as f
	inner join airports as a on f.departure_airport=a.airport_code
	inner join airports as aa on f.arrival_airport=aa.airport_code
	order by route)
select *, count(route) as '' 
from cte
except 
select *
from cte2;
```
